```python
import pandas as pd

movie_df = pd.read_csv("data/movies.dat", sep="::", header=0, names=["movie_id", "title", "genres"])
movie_df.head()
```

```python
genres_set = movie_df["genres"].str.split("|").explode().unique()
years_set = movie_df["title"].str.extract(r"\((\d{4})\)")[0].dropna().astype(int).unique()
len(genres_set), len(years_set), movie_df["movie_id"].nunique(), movie_df["title"].nunique()
```

```python
new_movie_df = movie_df.copy()
new_movie_df["year"] = new_movie_df["title"].str.extract(r"\((\d{4})\)")
new_movie_df["clean_title"] = new_movie_df["title"].str.replace(r"\(\d{4}\)", "", regex=True).str.strip()
new_movie_df.to_csv("processed/new_movies.csv", index=False)
new_movie_df.head()
```

```python
ratings_df = pd.read_csv("data/ratings.dat", sep="::", header=0, names=["user_id", "movie_id", "rating", "timestamp"])
ratings_df.to_csv("processed/new_ratings.csv", index=False)
ratings_df.head()
```

```python
users_df = pd.read_csv(
    "data/users.dat", sep="::", header=0, names=["user_id", "sex", "age", "occupation", "zip_code"]
)  # user_id::sex::age::occupation::zip_code
# users_df.to_csv('processed/new_users.csv', index=False)
# users_df.head()
users_df["age_bucket"] = users_df["age"] // 10
users_df.to_csv("processed/new_users.csv", index=False)
users_df.head()
```
